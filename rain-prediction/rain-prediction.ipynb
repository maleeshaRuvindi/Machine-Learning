```python
df_cleaned=df.dropna().copy()
```

```python
df_cleaned['RainTomorrow'].value_counts()
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df_cleaned[df_cleaned['RainTomorrow'] == 'No']
df_minority = df_cleaned[df_cleaned['RainTomorrow'] == 'Yes']

# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                   replace=False,        # without replacement
                                   n_samples=len(df_minority),  
                                   random_state=42)       # for reproducibility

# Combine minority and downsampled majority
df_undersampled = pd.concat([df_majority_downsampled, df_minority])

# Shuffle the combined dataset
df_undersampled = df_undersampled.sample(frac=1, random_state=42).reset_index(drop=True)

print(df_undersampled['RainTomorrow'].value_counts())
```

```python
df_undersampled.head()
```

```python
df_undersampled['RainTomorrow'] = df_undersampled['RainTomorrow'].map({'Yes': 1, 'No': 0})
```

```python
df_undersampled.columns
```

```python
features=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
       'Humidity3pm', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'RainToday'
       ]
X=df_undersampled[features].copy()
y=df_undersampled['RainTomorrow'].copy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
numerical_features=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
       'Humidity3pm', 'Pressure3pm', 'Cloud9am', 'Cloud3pm' 
       ]
categorical_features = ['RainToday']  # the column you want to leave as-is

# 1. Define the transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', 'passthrough', categorical_features)
    ]
)

# 2. Fit on training data and transform
X_train = preprocessor.fit_transform(X_train)

# 3. Transform test data
X_test = preprocessor.transform(X_test)
```

```python
learning_rate=0.01
w=np.zeros(X_train.shape[1])
n_rows=len(X_train)
w_history=[]
epochs=5000
e=[]
```

```python
for i in range(epochs):    
    z=np.dot(X_train,w.T)
    gz = 1/(1 + np.exp(-z)) 
    w=w-learning_rate*np.dot(X_train.T,gz-y_train)/n_rows
    e.append(sum(abs(y_train-gz))/n_rows)
    w_history.append(w.copy())


w_history=np.array(w_history)
```

```python
plt.plot(e)
plt.xlabel("Epoch")
plt.ylabel("Error")
plt.title("Error Updates over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python


for i in range(w_history.shape[1]):
    plt.plot([w[i] for w in w_history], label='w'+str(i))

plt.xlabel("Epoch")
plt.ylabel("Weight Value")
plt.title("Weight Updates over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Calculate final z on training data
z_final = np.dot(X_train, w.T)

# 2. Apply sigmoid to get probabilities
y_train_prob = 1 / (1 + np.exp(-z_final))

# 3. Convert probabilities to class predictions (threshold = 0.5)
y_train_pred = (y_train_prob >= 0.5).astype(int)

# 4. Check accuracy
from sklearn.metrics import accuracy_score

train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")
```

```python
# Predict on X_test
z_test = np.dot(X_test, w.T)
y_test_prob = 1 / (1 + np.exp(-z_test))
y_test_pred = (y_test_prob >= 0.5).astype(int)

# Test accuracy
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {test_accuracy:.4f}")
```
